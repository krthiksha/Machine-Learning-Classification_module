## Problem Statement  
# Social Network Ad's (digital marketing Domain)     --- Using Naive Bayes
### Predict : User will purchase the product or not
### if User purchase -> show ad
### if User won't purchase -> don't show the ad 

## Read the dataset 

```python
#import libraries
import pandas as pd 

# read the dataset
dataset = pd.read_csv("Social_Network_Ads.csv")
print(dataset)
```

## data Preprocessing 

```python
# convert categorial data into numerical data
dataset = pd.get_dummies(dataset, drop_first=True,dtype=int)
print(dataset)
```

```python
# Remove the unnecessary column (UserID) form the dataset   # axis = 1 (column)
dataset = dataset.drop(["User ID"], axis=1)
print(dataset)
```

## Check dataset is balanced or Imbalanced 

```python
# Check whether the dataset is Balanced or Imbalanced
"""
Purchased
0    257
1    143

Imbalanced dataset
"""
output_class_count = dataset["Purchased"].value_counts()
output_class_count
```

## Split input & output   |  Split train & test set 

```python
# split input & output
independent = dataset[['Age', 'EstimatedSalary','Gender_Male']]
dependent = dataset[['Purchased']]
print("Input : ", independent)
print("Output : ", dependent)
```

```python
# split train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
print(x_train,x_test,y_train,y_test )
```

## model creation | model Predicition | Model Evaluation

```python
y_train = y_train.values.ravel()
```

### GaussianNB 

```python
# model creation
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train,y_train)
# model prediction
y_pred = classifier.predict(x_test)
# model evaluation
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(y_test,y_pred)
print("Confusion Matrix: \n",matrix)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,y_pred)
print("\nClassification report: \n",clf_report)   # accuracy : 0.89  -- Good performance
```

### MultinomialNB 

```python
# model creation
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(x_train,y_train)
# model prediction
y_pred = classifier.predict(x_test)
# model evaluation
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(y_test,y_pred)
print("Confusion Matrix: \n",matrix)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,y_pred)
print("\nClassification report: \n",clf_report)   # accuracy : 0.68   -- Poor performance   
```

### ComplementNB

```python
# model creation
from sklearn.naive_bayes import ComplementNB
classifier = ComplementNB()
classifier.fit(x_train,y_train)
# model prediction
y_pred = classifier.predict(x_test)
# model evaluation
from sklearn.metrics import ( confusion_matrix, classification_report)
matrix = confusion_matrix(y_test,y_pred)
print("Confusion Matrix: \n",matrix)
clf_report = classification_report(y_test,y_pred)
print("\nClassification report: \n",clf_report)   # accuracy : 0.68  -- Poor performance
```

### BernoulliNB

```python
# model creation
from sklearn.naive_bayes import BernoulliNB
classifier = BernoulliNB()
classifier.fit(x_train,y_train)
# model prediction
y_pred = classifier.predict(x_test)
# model evaluation
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(y_test,y_pred)
print("Confusion Matrix: \n",matrix)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,y_pred, zero_division=0)   # zero_division=0 solves -- UndefinedMetricWarning
print("\nClassification report: \n",clf_report)   # accuracy : 0.66   -- Poor performance - weak model
```

### CategoricalNB

```python
# model creation
from sklearn.naive_bayes import CategoricalNB
classifier = CategoricalNB()
classifier.fit(x_train,y_train)
# model prediction
y_pred = classifier.predict(x_test)
# model evaluation
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(y_test,y_pred)
print("Confusion Matrix: \n",matrix)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,y_pred)
print("\nClassification report: \n",clf_report)   # accuracy :  0.90  -- Good performance
```
