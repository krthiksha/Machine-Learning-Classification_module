## Problem Statement  
# Social Network Ad's (digital marketing Domain)     --- Using Logistic Regression 
### Predict : User will purchase the product or not
### if User purchase -> show ad
### if User won't purchase -> don't show the ad 

## Read the dataset 

```python
#import libraries
import pandas as pd 

# read the dataset
dataset = pd.read_csv("Social_Network_Ads.csv")
print(dataset)
```

## data Preprocessing 

```python
# convert categorial data into numerical data
dataset = pd.get_dummies(dataset, drop_first=True,dtype=int)
print(dataset)
```

```python
# Remove the unnecessary column (UserID) form the dataset   # axis = 1 (column)
dataset = dataset.drop(["User ID"], axis=1)
print(dataset)
```

## Check dataset is balanced or Imbalanced 

```python
# Check whether the dataset is Balanced or Imbalanced
"""
Purchased
0    257
1    143

Imbalanced dataset
"""
output_class_count = dataset["Purchased"].value_counts()
output_class_count
```

## Split input & output   |  Split train & test set 

```python
# split input & output
independent = dataset[['Age', 'EstimatedSalary','Gender_Male']]
dependent = dataset[['Purchased']]
print("Input : ", independent)
print("Output : ", dependent)
```

```python
# split train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
print(x_train,x_test,y_train,y_test )
```

## model creation | model Predicition

```python
# solves DataConversionWarning - by converting Y as 1D array
y_train = y_train.values.ravel()
y_train
```

```python
# model creation 
from sklearn.linear_model import LogisticRegression
"""
max_iter=120 --> solved ConvergenceWarning by increaing max_iter which uses iterative optimization to find best model
solver{‘lbfgs’, ‘liblinear’, ‘newton-cg’, ‘newton-cholesky’, ‘sag’, ‘saga’}, default=’lbfgs’
"""
classifier = LogisticRegression(random_state = 0,max_iter=120 , C=1.0 , solver= 'lbfgs')   
classifier.fit(x_train, y_train)
```

```python
# model prediction
y_pred = classifier.predict(x_test)
y_pred
```

## Model Evaluation 

```python
# evaluate the model using confusion matrix
from sklearn.metrics import confusion_matrix 
test_set_matrix = confusion_matrix(y_test,y_pred)
print(test_set_matrix)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
display_labels = ["class 0","class 1"]
disp = ConfusionMatrixDisplay(confusion_matrix = test_set_matrix, display_labels=display_labels)
disp.plot()
plt.show()
```

```python
# classification report -- to find the derived metrics
from sklearn.metrics import classification_report 
clf_report = classification_report(y_test, y_pred)
print(clf_report)

"""
Algorithm : Logistic Regression 
Accuracy (overall performance of the model)  = 0.89
overall metrics performance (precision,recall,f1-score) = good 

Result : good model but not better than Randomforestclassifier
"""
```
