## Ridge regression   - regression problem (insurance charge prediction)

```python
# read the dataset 
import pandas as pd
dataset = pd.read_csv("insurance_pre.csv")
dataset.head()
```

```python
# convert the categorical data into numerical data 
dataset = pd.get_dummies(dataset, drop_first=True, dtype=int)
dataset.head()
```

```python
dataset.columns
```

```python
# split input and output 
independent = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes' ]]
dependent = dataset[[ 'charges']]
```

```python
dependent.head()
```

```python
independent.head()
```

```python
# split train & test 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(independent, dependent, test_size=0.20, random_state=0)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
# Standardization 
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

print("x_train: ",x_train, "\n\n x_test: ",x_test)
```

```python
# model creation 
from sklearn.linear_model import Ridge
regressor = Ridge(alpha=1.0)
regressor.fit(x_train,y_train)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
# model prediction
y_pred = regressor.predict(x_test)
y_pred[1:10] #1st 10 rows
```

```python
# evaluating the model 
from sklearn.metrics import r2_score
r2_score = r2_score(y_test,y_pred)
r2_score
```

## Ridge CV

```python
# model creation using --> cross validation in ridge -- ridge CV

from sklearn.linear_model import RidgeCV
alphas = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
ridgecv = RidgeCV(alphas=alphas, cv=5)    #choose the best alpha 
ridgecv.fit(x_train,y_train)
```

```python
# model prediction
y_pred_ridgecv = ridgecv.predict(x_test)
y_pred_ridgecv[1:10] #1st 10 rows
```

```python
# evaluating the model 
from sklearn.metrics import r2_score
r2_score = r2_score(y_test,y_pred_ridgecv)
r2_score
```

## check overt fit under fit  

```python

train_score = regressor.score(x_train,y_train) 
test_score = regressor.score(x_test,y_test) 
print(
    "train_score: ",train_score ,"\n",
    "test_score: ",test_score ,"\n",
    "train and test difference: ", abs(train_score - test_score),"\n"
)
#  train and test difference:  0.0616618957183096  -- good fit 
```

```python
train_score = ridgecv.score(x_train,y_train) 
test_score = ridgecv.score(x_test,y_test) 
print(
    "train_score: ",train_score ,"\n",
    "test_score: ",test_score ,"\n",
    "train and test difference: ", abs(train_score - test_score),"\n"
)

#  train and test difference:  0.0616618957183096  -- good fit 
```
